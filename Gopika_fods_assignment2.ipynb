```python
plt.figure(figsize=(8,6))
sns.boxplot(df['Total Profit'], color="orange")
plt.title('Boxplot for Total Profit')
plt.xlabel('Total Profit')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(df['Units Sold'], kde=True, color="blue", bins=15)
plt.title('Distribution of Units Sold')
plt.xlabel('Units Sold')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Total Profit'], kde=True, color="red", bins=15)
plt.title('Distribution of Total Profit')
plt.xlabel('Total Profit')
plt.ylabel('Frequency')
plt.show()
```
